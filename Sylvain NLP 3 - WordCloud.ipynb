```python
tokens = nltk.word_tokenize(texte.lower())
```

```python
tokens_clean = []
for words in tokens:
  if words not in nltk.corpus.stopwords.words("english"):
    tokens_clean.append(words)

tokens_clean
```

```python
# Remove punctuation
texte_sans = []
ponctuation = ".,;!?:'"
special = "'s"
guillemets = "«»"
for i in tokens_clean:
  if (i in ponctuation) or (i in special) or (i in guillemets):
  # Le caractère est présent dans la ponctuation
    pass
  else:
  # Le caractère n'est pas présent dans la ponctuation
    texte_sans.append(i)

print(texte_sans)
```

```python
new_texte = " ".join(texte_sans)
print(new_texte)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10)

# Génération du wordcloud depuis un texte
wordcloud.generate_from_text(new_texte)

# Affichage grâce à Matplotlib
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
