3. Crée une fonction que tu appelleras clean qui prend pour paramètre une phrase (donc un texte str) et qui retourne un texte (str) de tokens après application d'un stemmer ou d'un lemmatizer, séparés par des espaces

```python
from nltk.stem import SnowballStemmer
```

```python
phrase = "You are better when I am well."
```

```python
def clean(phrase):
  stem_en = SnowballStemmer("english")
  for word in nltk.word_tokenize(phrase):
    print(stem_en.stem(word))
```

```python
clean("You are better when I am well.")
```

```python
# Conversion du texte en minuscule
texte_minuscule = phrase.lower()

from string import punctuation

def remove_punctuation_nltk(texte_minuscule):
  """Removes punctuation from text using NLTK's translate() function."""
  translation_table = str.maketrans('', '', punctuation)
  return texte_minuscule.translate(translation_table)
```

```python
tokens_clean = []
for words in texte_minuscule:
  if words not in nltk.corpus.stopwords.words("english"):
    tokens_clean.append(words)

tokens_clean
```

```python
# Remove punctuation
from string import punctuation

tokens_without_punctuation = []
for word in tokens_clean:
  if word not in punctuation:
    tokens_without_punctuation.append(word)

# Print the cleaned tokens
print(tokens_without_punctuation)
```
